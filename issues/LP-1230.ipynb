```python
import pandas as pd
import numpy as np
from lets_plot import *
```

```python
LetsPlot.setup_html()
```

```python
data = {
    'category': ['This is a label', 'Label B', 'C', 'D', 'This is a label E'],
    'value': [5, 4, 3, 2, 1]
}
train = pd.DataFrame(data)
```

```python
for i in [0, 0.5, 1]:
    plot = (
        ggplot(train, aes(x='category', y='value'))
        + geom_bar(stat='identity')
        + ggtitle(f'Sample plot: vjust & hjust = {i}')
        + theme(axis_text_x=element_text(angle=45, hjust=i, vjust=i))
    )
    plot.show()
```

```python
td = pd.DataFrame({
    'hjust': np.tile([0, 0.5, 1], 9),
    'vjust': np.tile([0, 0.5, 1], 9),
    'angle': np.repeat([0, 45, 90], 9),
    'text': ["text"] * 27
})
```

```python
ggplot(td, aes(x='hjust', y='vjust')) + \
    geom_point() + \
    geom_text(aes(label='text', angle='angle', hjust='hjust', vjust='vjust')) + \
    facet_grid(x='angle') 
```
