```r
library(ggplot2)
```

```r

# Prepare the data
data <- data.frame(
  category = c('This is a label', 'Label B', 'C', 'D', 'This is a label E'),
  value = c(5, 4, 3, 2, 1)
)
```

```r
for (i in c(0, 0.5, 1)) {
  plot <- ggplot(data, aes(x = category, y = value)) +
    geom_bar(stat = 'identity', fill = 'blue') +
    ggtitle(paste('Sample plot: vjust & hjust =', i)) +
    theme(axis.text.x = element_text(angle = 45, hjust = i, vjust = i))
  
  print(plot)
}
```
